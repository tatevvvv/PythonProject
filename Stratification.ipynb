```python
import pandas as pd

# Creating the DataFrame
data = pd.DataFrame(
    [
        ["lung", 40, 60, 100],       # Row for Smokers
        ["no-lung", 10, 90, 100]   # Row for Non-Smokers
    ],
    columns=["status", "smokers", "non-smokers", "total"]  # Column names
)

# Display the DataFrame
print(data)
```

```python
# Calculate the crude odds ratio to see if there is an association between smoking and lung disease without accounting for any confounders.
smokersWhoHasCancer = data['smokers'][0]
NonSmokersWhoHasCancer = data['non-smokers'][1]

SmokersWithOutCancer = data['smokers'][1]
NonSmokersWithoutCancer = data['non-smokers'][1]

# Calculate risk for smokers and non-smokers
smokersRisk = smokersWhoHasCancer / NonSmokersWhoHasCancer
nonSmokersRisk = SmokersWithOutCancer / NonSmokersWithoutCancer

print(f"Risk for Smokers: {smokersRisk}")
print(f"Risk for Non-Smokers: {nonSmokersRisk}")

print(f"Crude odds for Smoking and Lung Cancer {smokersRisk / nonSmokersRisk}")
```
